# K-means

determine the optimale number of clusters


1.   elbow methode
2.   silhouette methode




````r
# Calculate total sum of squares (total_SS)
total_SS <- sum((data[-8] - colMeans(data[-8]))^2)

# Compute total within-cluster sum of squares (WCSS) for different values of k
wcss <- numeric(10)
for (k in 2:10) {
  kmeans_result <- kmeans(data[-8], centers = k)
  wcss[k] <- kmeans_result$tot.withinss
}

# Plot the elbow evaluation
plot(2:10, wcss[2:10], type = "b", xlab = "Number of clusters (k)", ylab = "Proportion of Variance Explained",
     main = "Elbow Method for Optimal Number of Clusters", col = "blue", pch = 19, lwd = 2, cex.axis = 1.5, cex.lab = 1.5, cex.main = 1.5)

# Calculate the slope and intercept for the extended line
a <- (wcss[5] - wcss[4]) / (5 - 4)
b <- wcss[5] - a * 5
line_values <- a * 2:10 + b

# Add the extended line
lines(2:10, line_values, col = "red", lty = 2)
````

````r
# Compute differences between successive values of WCSS
diff_wcss <- diff(wcss)
print(diff_wcss)

# Create the stair plot
plot(2:10, diff_wcss[2:10], type = "s", xlab = "Number of clusters (k)",
     ylab = "Difference in WCSS", main = "Difference in WCSS between successive cluster numbers",
     col = "blue", lwd = 2, cex.axis = 1.5, cex.lab = 1.5, cex.main = 1.5)
````

````r
library(cluster)

# Compute the distance matrix
distance_matrix <- dist(data[-8])

silhouette_scores <- c()
silhouette_scores[1] <- 0

for (k in 2:10) {
  kmeans_result <- kmeans(data[-8], centers = k)
  cluster_assignments <- kmeans_result$cluster
  sil <- silhouette(cluster_assignments, distance_matrix)
  silhouette_scores[k] <- mean(sil[, "sil_width"])
}

# Find the index of the maximum silhouette score
max_index <- which.max(silhouette_scores)
optimal_clusters <- max_index

# Plot the silhouette scores with customizations
plot(1:10, silhouette_scores, type = "b", xlab = "Number of clusters (k)", ylab = "Average Silhouette Score",
     main = "Silhouette Analysis for Optimal Number of Clusters", ylim = c(0, 1),
     col = "black", pch = 19, lwd = 2, cex.axis = 1.5, cex.lab = 1.5, cex.main = 1.5)

# Add a dashed vertical line at the optimal number of clusters
abline(v = max_index, lty = 2, col = "red")
````

````r
# Find the index of the maximum silhouette score
max_index <- which.max(silhouette_scores)

cat("Optimal number of clusters:", optimal_clusters, "\n")
cat("Maximum silhouette score:", silhouette_scores[max_index], "\n")
````

````r
# Fonction pour obtenir les corrélations triées par magnitude
sortCorrelations <- function(cor_matrix) {
  cor_df <- as.data.frame(as.table(cor_matrix))
  colnames(cor_df) <- c("Var1", "Var2", "Correlation")

  cor_df$Var1 <- as.character(cor_df$Var1)
  cor_df$Var2 <- as.character(cor_df$Var2)

  cor_df <- cor_df[cor_df$Var1 != cor_df$Var2, ]
  cor_df <- cor_df[cor_df$Var1 < cor_df$Var2, ]

  # Trier par valeurs absolues de corrélation
  cor_df <- cor_df[order(abs(cor_df$Correlation), decreasing = TRUE), ]

  rownames(cor_df) <- NULL

  return(cor_df)
}

sorted_correlations <- sortCorrelations(data.correlations)

# Affichage les corrélations triées
print(sorted_correlations)
````

# CAH


**Introduction**

Dans cette section, nous appliquons la Classification Ascendante Hiérarchique (CAH) pour découvrir des groupements naturels au sein de notre jeu de données de véhicules. Nous débutons par la normalisation des données pour équilibrer l'échelle des variables, suivie de l'application de la méthode de Ward pour identifier les clusters. L'objectif est d'interpréter ces regroupements à travers des visualisations et des analyses de correspondance pour mieux comprendre les patterns sous-jacents.

````r
# Normalisation des données
data_scaled <- scale(data[-8])

# Calcul de la matrice de distance euclidienne
data_dist <- dist(data_scaled)

# Application de la méthode de Ward
data_hc_ward <- hclust(data_dist, method = "ward.D2")

# Visualisation du dendrogramme
plot(data_hc_ward, main = "Dendrogramme de la CAH", xlab = "Observations", ylab = "Distance")
````

**Analyse de l'inertie:**

Nous déterminons le nombre optimal de clusters en analysant l'inertie, ce qui nous aide à comprendre jusqu'où les clusters sont distincts et compacts.

````r
# Calcul de l'inertie
inertia <- sort(data_hc_ward$height, decreasing = TRUE)

# Visualisation de l'inertie
plot(inertia[1:20], type = "s", xlab = "Nombre de clusters", ylab = "Inertie", main = "Graphique de l'inertie")
````

**Déterminer le nombre de clusters**

Nous utilisons la méthode du coude pour sélectionner le nombre optimal de clusters.



````r
# Identifier le point de "coude" pour déterminer le nombre optimal de clusters
elbow <- which(diff(inertia) == max(diff(inertia)))
num_clusters <- elbow
print(paste("Nombre de clusters optimal (coude):", elbow))
````

**Attribution des Clusters aux Données et Visualisation sue le Dendrogrmme**

Nous coupons le dendrogramme au niveau du nombre de clusters choisi et ajoutons une nouvelle colonne aux données contenant les étiquettes de cluster.



````r
# Application du nombre optimal de clusters
cluster_labels <- cutree(data_hc_ward, num_clusters)
# Ajout des étiquettes de clusters aux données
data$cluster <- cluster_labels
# Affichage des étiquettes de clusters
print(data$cluster)
````

````r
# Visualisation du dendrogramme avec les partitions proposées
par(mar=c(1,4.5,2,0))
plot(data_hc_ward, labels = FALSE, main = "Partitions en 2, 3 ou 4 classes", xlab = "", ylab = "Hauteur", sub = "", hang = -1)
rect.hclust(data_hc_ward, 2, border = "green")
rect.hclust(data_hc_ward, 3, border = "red")
rect.hclust(data_hc_ward, 4, border = "blue")
````

**Analyse des Clusters**

Nous analysons les clusters en examinant les tables de contingence.

````r
# Fonction pour afficher les tables de contingence avec pourcentages
afficher_table_contingence_clusters <- function(factor, clusters) {
    contingency_table <- table(factor, clusters)
    percentages <- prop.table(contingency_table, margin = 1) * 100
    print(round(percentages, 2))
}

# Utilisation de la fonction pour différentes découpes
cut_affiche_table_contingence = function(hc_result, cut_vector, factor) {
    for(cut in cut_vector) {
        cat(paste("\nCut =", cut, "\n"))
        c = cutree(hc_result, cut)
        afficher_table_contingence_clusters(factor, c)
    }
}

# Affichage des tables de contingence pour les différents clusters
cut_affiche_table_contingence(data_hc_ward, c(2, 3, 4), data$brand)
````

````r
# Tracer l'inertie du dendrogramme avec les plus grands gains entourés
plot(inertia[1:20], type = "s", xlab = "Nombre de clusters", ylab = "Inertie", main = "Inertie du dendrogramme avec plus grands gains entourés")
points(c(2, 3, 4), inertia[c(2, 3, 4)], col = c("green", "red", "blue"), cex = 2, lwd = 3)
````
